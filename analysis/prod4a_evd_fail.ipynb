```python
%matplotlib qt5
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from tabulate import tabulate
import awkward as ak
import numpy as np

import Master
import vector

from prod4a_merge_study import EventSelection, ShowerMergeQuantities, Percentage
from prod4a_evd import EventDisplay
    
```

```python
events = Master.Data("work/ROOTFiles/Prod4a_6GeV_BeamSim_00_evd.root", True)
events.recoParticles.spacePoints = ak.zip({"x" : events.io.Get("reco_daughter_allShower_spacePointX"), 
                                           "y" : events.io.Get("reco_daughter_allShower_spacePointY"),
                                           "z" : events.io.Get("reco_daughter_allShower_spacePointZ")})
events.Filter([events.recoParticles.startPos.x != -999])
EventSelection(events, invertFinal=True)
```

```python
def CNNFilter():
    return events.recoParticles.cnnScore[i] > 0.64
def HitsFilter():
    return events.recoParticles.nHits[i] > 0
def BeamDaughterFilter():
    return events.recoParticles.mother[i] == events.recoParticles.beam_number[i]

i = 0
display = EventDisplay(events.eventNum[i], events.run[i], events.subRun[i])

filter = np.logical_and(CNNFilter(), HitsFilter())

nPFO = ak.count(events.recoParticles.nHits[i][filter])
print(f"nPFO: {nPFO}")
for j in range(nPFO):
    if events.recoParticles.number[i][j] == events.recoParticles.beam_number[i]: continue
    display.PlotPFO(events.recoParticles.spacePoints[i][filter][j], marker="x", startPoint=events.recoParticles.startPos[i][filter][j], colour=f"C{j}")#, pdg=events.trueParticlesBT.pdg[i][filter][j])

print(events.trueParticles.mother[i])
print(events.trueParticles.number[i])

beam = events.recoParticles.number[i] == events.recoParticles.beam_number[i]
print(events.trueParticlesBT.number[i][beam][0])

display.PlotPFO(events.recoParticles.spacePoints[i][beam], marker="x", startPoint=events.recoParticles.startPos[i][beam], colour="black")

colours = ["red", "blue", "green"]
for j in range(3):
    display.PlotPoint(events.trueParticles.startPos[i][j], "o", colours[j], pointSize=16)
    display.PlotText(events.trueParticles.endPos[i][j], events.trueParticles.pdg[i][j])
    display.PlotLine(events.trueParticles.startPos[i][j], events.trueParticles.endPos[i][j], colours[j])

```
