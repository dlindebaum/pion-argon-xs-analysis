```python
from python.analysis import Master, Plots
from apps.prod4a_merge_study import EventSelection, PFOSelection, SplitSample

import awkward as ak
import pandas as pd
import matplotlib.pyplot as plt

from rich import print

plt.style.use('ggplot')
plt.rcParams.update({'patch.linewidth': 1})
plt.rcParams.update({'font.size': 10})
```

```python
events = Master.Data("work/ROOTFiles/Prod4a_1GeV_BeamSim_00_test_hits.root")

EventSelection(events)
PFOSelection(events)
start_showers, to_merge = SplitSample(events)

merge = events.Filter(returnCopy = True)
merge.MergePFOCheat(1)

data = []
for i in range(2):
    mask = merge.trueParticlesBT.number == ak.flatten(events.trueParticlesBT.number[start_showers[i]])
    data.append(pd.DataFrame({
        "true hits" : ak.ravel(events.trueParticlesBT.nHits_collection[start_showers[i]]),
        "start shower hits" : ak.ravel(events.recoParticles.nHits[start_showers[i]]),
        "merged shower hits" : ak.ravel(merge.recoParticles.nHits[mask]),
        "start shower shared hits" : ak.ravel(events.trueParticlesBT.sharedHits_collection[start_showers[i]]),
        "merged shower shared hits" : ak.ravel(merge.trueParticlesBT.sharedHits_collection[mask]),
        "start shower completeness" : ak.ravel(events.trueParticlesBT.completeness_collection[start_showers[i]]),
        "merged shower completeness" : ak.ravel(merge.trueParticlesBT.completeness_collection[mask])
    }))
data = pd.concat(data).reset_index()
data = pd.concat([data, pd.DataFrame({"missing hits" : (data["true hits"] - data["merged shower hits"]).values})], axis = 1)
data = pd.concat([data, pd.DataFrame({"nHits merged" : (data["merged shower hits"] - data["start shower hits"]).values})], axis = 1)

for k in data.columns:
    Plots.PlotHist(data[k], xlabel = k, bins = 50)
    Plots.Save(k, "test/missing_hits/")
```
